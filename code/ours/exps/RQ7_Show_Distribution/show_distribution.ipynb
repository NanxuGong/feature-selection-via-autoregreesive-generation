```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
def plot(Dg):
	X = Dg.iloc[:, :-1]
	y = Dg.iloc[:, -1]
	tsne = TSNE(n_components=2)
	X_tsne = tsne.fit_transform(X)

	plt.figure(figsize=(7, 5), dpi=300)
	plt.scatter(X_tsne[:, 0], X_tsne[:, 1], s=5, c=y, cmap=plt.cm.Spectral)
	plt.show()
```

```python
plot(ours)
```

```python
plot(rfe)
```

```python
plot(original)
```
