```python
import matplotlib.pyplot as plt
import seaborn as sns
import palettable#python颜色库
```

```python
def show_img(name):
	ds = pd.read_csv(f'/home/xiaomeng/jupyter_base/AutoFS/code/ours/exps/RQ5:More_Downstream/out/{name}.csv')
	df = ds.iloc[:, 1:]
	df.index = ds.iloc[:, 0]
	df = df * 100
	plt.figure(dpi=400)
	ax = sns.heatmap(data=df, cmap=plt.get_cmap('Greens_r'),annot=True,fmt='.1f')
	ax.set(xlabel = 'Feature Selection Method', ylabel = 'Downstream Model', title=name)
	plt.show()
```

```python
ds = pd.read_csv('/home/xiaomeng/jupyter_base/AutoFS/code/ours/exps/RQ5:More_Downstream/out/german_credit.csv')
df = ds.iloc[:, 1:]
df.index = ds.iloc[:, 0]
df = df * 100
df.transpose
plt.figure(dpi=400,figsize=(5,5))
sns.set(font_scale=1.0)
ax = sns.heatmap(data=df.transpose(), linewidths=0.0,cmap="rocket_r",annot=True, annot_kws={"fontsize":11},fmt='.1f', square=False)
ax.set_xlabel('Downstream Model', fontsize=12)
ax.set_ylabel('Feature Selection Method', fontsize=12)
# ax.set_title('German Credit', fontsize=16)
plt.xticks()
plt.savefig('/home/xiaomeng/jupyter_base/AutoFS/code/ours/exps/RQ5:More_Downstream/out/german_ensemble.png', bbox_inches='tight')
plt.show()
```
